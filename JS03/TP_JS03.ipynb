```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

df = pd.read_csv("wbc.csv")
#1
df = df.drop(columns=["id", "Unnamed: 32"], errors="ignore")
#2
label_enc = LabelEncoder()
df["diagnosis"] = label_enc.fit_transform(df["diagnosis"])
#3
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]
# 4
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("select", SelectKBest(score_func=f_classif)),
    ("clf", LogisticRegression(max_iter=500))
])
# 5
feature_range = range(5, X.shape[1] + 1)
scores = []

for k in feature_range:
    pipeline.set_params(select__k=k)
    cv_scores = cross_val_score(pipeline, X, y, cv=5, scoring="accuracy")
    scores.append(cv_scores.mean())

best_k = feature_range[int(np.argmax(scores))]
best_score = max(scores)

print("Jumlah fitur terbaik:", best_k)
print("Akurasi terbaik (CV 5-fold):", best_score)
# 6
pipeline.set_params(select__k=best_k)
pipeline.fit(X, y)
# 7
selector = pipeline.named_steps["select"]
mask = selector.get_support()
best_features = X.columns[mask]
print("\nFitur-fitur terbaik:")
print(best_features.tolist())
```
