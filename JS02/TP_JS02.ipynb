```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Upload file wbc.csv via panel kiri “Files” di Colab
df = pd.read_csv('wbc.csv')
df.head()
```

```python
print("Semua kolom:", df.columns.tolist())

# Buang kolom yang tidak relevan (contoh: id, unnamed)
drop_cols = [col for col in df.columns if 'id' in col.lower() or 'unnamed' in col.lower()]
df_used = df.drop(columns=drop_cols)

print("Kolom yang digunakan:", df_used.columns.tolist())
print("Kolom yang dibuang:", drop_cols)
```

```python
le = LabelEncoder()
df_used['diagnosis'] = le.fit_transform(df_used['diagnosis'])
# Pastikan mappingnya:
mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Mapping diagnosis:", mapping)
df_used.head()
```

```python
num_cols = df_used.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
df_used[num_cols] = scaler.fit_transform(df_used[num_cols])

# Tampilkan beberapa baris hasil:
df_used.head()
df_used.describe()
```
